## Data visualization with Seaborn
Data visualization is an important stage in the data mining process because it helps us better understand the patterns in the data. Even after fitting a model to the data set, it can be useful to go back to the data exploration stage in order to refine one's understanding of the model output.

There are several Python libraries available for data visualization with Python. The most popular are [Matplotlib](https://matplotlib.org/) and [Seaborn](https://seaborn.pydata.org/), but the [Pandas](https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html) library has several data visualization methods built in. In these exercises, we will explore a few useful graphics that can be created with Seaborn.



```python
# Read in the data set (built into Seaborn, we just have to import it)
import seaborn as sns
df = sns.load_dataset("tips")

# Examine the first few lines
df.head()
```

Example 1: Creating a bar graph to look at **tips** by **day** of the week.


```python
ax = sns.barplot(x="day", y="tip", data=df,
                 palette="Blues_d", errorbar=None)
```

Exercise 1: Creating a bar graph to look at **total bill** by party **size**.

```python
ax = sns.barplot(x="size", y="total_bill", data=df,
                 palette="Blues_d", errorbar=None)
```

Example 2: Creating a stacked bar graph to visualize the count of parties by **day** of the week and party **size**.

```python
# One liner to create a stacked bar chart.
ax=sns.histplot(df,
                x='day',
                hue='size',
                multiple='stack',
                palette='tab20c',
                shrink=0.8)
ax.set_title('Party Size by Weekday')
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

Exercise 2: Creating a stacked bar graph to visualize the count of parties by **day** of the week and **smoking** status.

```python
# One liner to create a stacked bar chart.
ax=sns.histplot(df,
                x='day',
                hue='smoker',
                multiple='stack',
                palette='tab20c',
                shrink=0.8)
```

Example 3: Creating a Histogram (numerical variable) of the **total bill**. See [color list](https://matplotlib.org/stable/gallery/color/named_colors.html)

```python
ax = sns.histplot(df, x='total_bill', stat='percent', color='burlywood')
```

Exercise 3: Creating a Histogram (numerical variable) of the **tips**. See [color list](https://matplotlib.org/stable/gallery/color/named_colors.html)

```python
ax = sns.histplot(df, x='tip', stat='percent', color='darkkhaki')
```

Example 4: Creating a Scatterplot. For the full collection of color options see [Color Palettes in Seaborn](https://medium.com/@morganjonesartist/color-guide-to-seaborn-palettes-da849406d44f). For more examples of legend positioning see [this page](https://www.statology.org/seaborn-legend-position/).

```python
# build the scatterplot
ax=sns.scatterplot(data=df,
                   x='total_bill',
                   y='tip',
                   hue='day',
                   size='size',
                   palette='RdBu')

# customize the legend position
ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```
